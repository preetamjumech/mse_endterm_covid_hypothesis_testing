# Z Test for Proportion

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
tab1.loc["All":]
```

```python
cal_tab=tab1.loc[["india","china"],["delta","All"]]
cal_tab
```

```python
def prop_test_registered(c):
    print(c)
    stat, pval = proportions_ztest(cal_tab["delta"],cal_tab["All"])
    print(stat)
    print(pval)
    print("  ")
```

```python
prop_test_registered(cal_tab)
```

## p value is 0.6732. Hence there is no significant difference in propertions of delta in india and china

# TTEST

```python
def ttest(col):
    print(f"FOR {col}")
    delta_no_days_before_recover=df[df.variant=="delta"]["no_days_before_recover"]
    omicorn_no_days_before_recover=df[df.variant=="omicorn"]["no_days_before_recover"]
    print(delta_no_days_before_recover.values)
    print(omicorn_no_days_before_recover.values)
    print(ttest_ind(delta_no_days_before_recover.values,omicorn_no_days_before_recover.values))
```

```python
ttest("variant")
```

## p value is 0.8588.Hence there is no significant difference in no_days_before_recover between delta and omicorn

# CHI SQUARE TEST

```python
def chi_square(crosstab):
    stat, p, dof, expected = chi2_contingency(crosstab)
    return(p)
```

```python
ct=pd.crosstab(df["country"],df["variant"])
print(ct)
```

```python
print(chi_square(ct))
```

## p value is 0.1498.Hence the country and variant(omicorn and delta) have a significant association between them

```python
def regression_results(X,y):
    x=sm.add_constant(X)
    model=sm.OLS(y,x).fit()
    print(model.summary())
```
